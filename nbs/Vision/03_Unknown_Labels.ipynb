```python
from fastai2.vision.all import *
```

We'll go ahead and make the dataloaders with one important change

```python
path = untar_data(URLs.PETS)/'images'
```

MultiCategory expects a list of labels

```python
def multi_l(l): return [l]
```

Now we pass that into our `get_y` along with any labeller

```python
pets_multi = DataBlock(blocks=(ImageBlock, MultiCategoryBlock),
                 get_items=get_image_files,
                 splitter=RandomSplitter(),
                 get_y=Pipeline([RegexLabeller(pat = r'/([^/]+)_\d+.jpg$'), multi_l]),
                 item_tfms=RandomResizedCrop(460, min_scale=0.75),
                 batch_tfms=[*aug_transforms(size=224, max_warp=0), Normalize.from_stats(*imagenet_stats)])
```

```python
dls = pets_multi.dataloaders(untar_data(URLs.PETS)/"images",  bs=32)
```

```python
dls.show_batch(max_n=9, figsize=(6,7))
```

## Model

I'm choosing a very high threshold for our metrics as we want only super confident answers (as we only have one label)

```python
learn = cnn_learner(dls, resnet34, pretrained=True, metrics=[partial(accuracy_multi, thresh=0.1)], 
                    loss_func=BCEWithLogitsLossFlat())
```

```python
learn.fit_one_cycle(4)
```

```python
learn.save('stage_1')
```

Let's see how we did!

```python
learn.show_results()
```
