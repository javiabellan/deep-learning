# MULTIPLE Random normal distribution

```
mean, std = 0, 0

N = 1000
for i in range(N):
  x = torch.randn(512)
  mean += x.mean().item()
  std  += x.std().item()
mean/N, std/N
```

# MULTIPLE Random normal distribution multiplication

```
mean, std, var = 0., 0., 0.

N = 10000
for i in range(N):
  x = torch.randn(512)
  w = torch.randn(512, 512)
  y = x @ w
  mean += y.mean().item()
  std  += y.std().item()
  var  += y.pow(2).mean().item()
mean/N, std/N, math.sqrt(var/N)
```

# Exploding Output (100 layers)
- Imagine no activation layers
- At layer 16, gradient explodes

```
x = torch.randn(512) # Input: Normal distr
print("Std at input:\t\t", x.std().item())

for i in range(100):
  w = torch.randn(512, 512)  # Weight: Normal distr
  x = w @ x                 # Output of layer
  if x.std()>10000000:      # Too big outputs
    print("At layer", i, "output explodes")
    break
  else:
    print("Std at layer", i, ":\t", x.std().item())
```

# Vanishing Output (100 layers)
- Imagine no activation layers
- At layer 16, gradient vanish

```
x = torch.randn(512) # Input: Normal distr
print("Std at input:\t\t", x.std().item())

for i in range(100):
  w = torch.randn(512, 512) * 0.01 # Weight: Normal distr
  x = w @ x                        # Output of layer
  if x.std()<0.000001:             # Too small outputs (NaN values)
    print("At layer", i, "output vanish")
    break
  else:
    print("Std at layer", i, ":\t", x.std().item())
```
