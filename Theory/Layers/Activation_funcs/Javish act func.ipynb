# Kaiming

```python
mean_relu,   std_relu = 0., 0.
mean_swish,  std_swish = 0., 0.
mean_mish,   std_mish = 0., 0.
mean_javish, std_javish = 0., 0.

N=1000
for i in range(N):
    x = torch.randn(512)
    w = kaiming(512, 512)
    x = x @ w
    
    x_relu   = relu(x)
    x_swish  = swish(x)
    x_mish   = mish(x)
    x_javish = javish(x)

    mean_relu   += x_relu.mean().item();   std_relu    += x_relu.std().item()
    mean_swish  += x_swish.mean().item();  std_swish   += x_swish.std().item()
    mean_mish   += x_mish.mean().item();   std_mish    += x_mish.std().item()
    mean_javish += x_javish.mean().item(); std_javish  += x_javish.std().item()

print("Relu mean:  ", mean_relu/N,   "std:", std_relu/N)
print("Swish mean: ", mean_swish/N,  "std:", std_swish/N)
print("Mish mean:  ", mean_mish/N,   "std:", std_mish/N)
print("Javish mean:", mean_javish/N, "std:", std_javish/N)
```
